```python
test = np.genfromtxt('C:\\Users\\snaug\\Downloads\\0000000000002329_test_diabetes_x_test.csv', delimiter=',')
x_test = pd.DataFrame(test)
x_test.columns = ['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']
x_test
```

```python
# predicting for Y values

y_predict = alg.predict(x_test)
```

```python
# calculating score or coefficient of determination

alg.score(x_train , y_train)
```
