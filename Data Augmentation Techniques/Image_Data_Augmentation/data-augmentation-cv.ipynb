## Image Data Augmentation Techniques

```python
# Import libraries
from keras.preprocessing.image import ImageDataGenerator
```

```python
# re-size all the images to this
IMAGE_SIZE = [224, 224]
```

```python
# Splitting the data into training(80%) and validation(20%) set using validation_split of ImageDataGenerator
# rescaling,shearing,zooming,flipping
data_datagen = ImageDataGenerator(validation_split=0.2,
                                   rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)


# Generating training_set from data_datagen using subset as training
training_set = data_datagen.flow_from_directory('../input/animal-image-dataset-90-different-animals/animals/animals/',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical',
                                                 color_mode="rgb", # for coloured images
                                                 subset="training")
# Generating valid_set from data_datagen using subset as validation
valid_set = data_datagen.flow_from_directory(
    '../input/animal-image-dataset-90-different-animals/animals/animals',
    target_size = (224, 224),
    batch_size = 32,
    class_mode = 'categorical',
    subset="validation"
)

```
